# Histograms

Distribution of Height for age Z-score ( HAZ), weight for age Z-score( WAZ) and weight for height z-score ( WHZ) for the children aged under 5 in Bangladesh.

```python
import seaborn as sns
plt.style.use('ggplot') # select a style (theme) for plot
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
sns.distplot(data[data['HAZ_child']>-10] ['HAZ_child'], kde=True, rug=False)
plt.title('Height for Age Z-score')

plt.subplot(1,3,2)
sns.distplot(data[data['WAZ_child']>-10 ]['WAZ_child'], kde=True, rug=False)
plt.title('Weight for Age Z-score')

plt.subplot(1,3,3)
sns.distplot(data[data['WHZ_child']>-10 ]['WHZ_child'], kde=True, rug=False)
plt.title('Weight for Height Z-score')

```

# Scatter Plots
To visualize correlations between variables, scatter plots can be quite useful.

Let's use a scatter plot to explore the correlation between mothers years of shooling  and child nutrition. 

```python
plt.figure(figsize=(8,6))
sns.lmplot(x='education_yrs_mother', y='HAZ_child', data=data , markers=['o'], 
           scatter_kws={'s':50, 'edgecolor':'Black'})
plt.title('Mothers education and child nutrition')
```

# Barplot
Bar plots are very useful to show the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.
Lets use a barplot to show the mean height for age-Z score of the children under-5 in each weath group of the households.


```python
sns.barplot(x='wealthgroup',y='HAZ_child', data=data,estimator=np.mean )
plt.title('Household wealth and child nutrition')
```
